```python
time_stamps = pd.read_pickle('pd_time_stamps.pkl')
time_stamps.head()
```

```python
def interpolate_to_timestamps(df):
    df = df.resample('4T').interpolate()
    ind = [df.index.get_loc(tim, method='nearest') for tim in time_stamps.time]
    return df.iloc[ind, :]
```

```python
df = df.set_index(pd.DatetimeIndex(df.loc[:, 'time']))
```

```python
df_interp = interpolate_to_timestamps(df)
```

```python
len(time_stamps)
```

```python
df2 = df2.set_index(pd.DatetimeIndex(df2.loc[:, 'time']))
df2_interp = interpolate_to_timestamps(df2)
```

```python
df3 = df3.set_index(pd.DatetimeIndex(df3.loc[:, 'time']))
df3_interp = interpolate_to_timestamps(df3)
```

```python
plt.scatter(df_interp.iloc[summer, :].loc[:, ['resistivity_10cm [Mohm]']], df2_interp.iloc[summer, :].loc[:, 'rain_intensity [mm/h]'], alpha=0.6)
```

```python
df_interp.time = df_interp.index
df_interp.plot('time', ['resistivity_10cm [Mohm]', 'resistivity_35cm [Mohm]', 'resistivity_60cm [Mohm]', 'resistivity_85cm [Mohm]'], figsize=(20,10))
plt.semilogy()
```
