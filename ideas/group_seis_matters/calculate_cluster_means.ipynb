```python
import numpy as np
import os
import matplotlib.pyplot as plt
file_path = "./"
```

```python
cluster1_path= "./ResNet50-withoutPCA/output\cluster0/"
cluster2_path= "./ResNet50-withoutPCA/output\cluster1/"
cluster3_path= "./ResNet50-withoutPCA/output\cluster2/"
cluster4_path= "./ResNet50-withoutPCA/output\cluster3/"
```

```python
cluster1_images= os.listdir(cluster1_path)
cluster2_images= os.listdir(cluster2_path)
cluster3_images= os.listdir(cluster3_path)
cluster4_images= os.listdir(cluster4_path)
```

```python
cl1= []
cl2= []
cl3=[]
cl4= []
```

```python
for im in cluster1_images:
    cl1.append(plt.imread(cluster1_path + im))
```

```python
for im in cluster2_images:
    cl2.append(plt.imread(cluster2_path + im))
for im in cluster3_images:
    cl3.append(plt.imread(cluster3_path + im))
for im in cluster4_images:
    cl4.append(plt.imread(cluster4_path + im))
```

```python
cl1_mean= np.mean(cl1, axis=0) / 255
cl2_mean= np.mean(cl2, axis=0) / 255
cl3_mean= np.mean(cl3, axis=0) / 255
cl4_mean= np.mean(cl4, axis=0) / 255
```
